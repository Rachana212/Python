# **Bar Plot**

A bar plot is a visualization used to represent categorical data with rectangular bar.The length or height of each bar is proportional to the value it represents .Bar plot are widely used for comparing quantities across categories ,making them a fundamental tool in data analysis and visualization


```python
df.groupby('day')['tip'].mean()
```

```python

import matplotlib.pyplot as plt
sns.barplot(data=df,x='day',y='tip',estimator='mean',errorbar=None,palette='Set1')
plt.show()
```

```python

ax=sns.barplot(data=df,x='day',y='tip',estimator='mean',errorbar=None)
ax.bar_label(ax.containers[0],fontsize=10)
plt.grid()
plt.show()
```

```python
df.groupby('day')['tip'].mean().values
```

```python

ax= sns.barplot(data=df, x='day', y='tip', estimator='mean', errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)
plt.grid()
plt.xlabel('days')
plt.ylabel('Avg tip')
plt.title('Avg tips by days')
plt.show()
```

# **Scatter Plot**

A scatter plot isa type of visualization used to dusplay the relationship b/w 2 numerical variable .Ecah point in the scatter plot represent an observation,with its position determined by the values of two variables -one on the x-axis and the other y-axis.

# **Box Plot**

A box plot is a standardized way of displaying the distribution of data on a five number summary,(1st,2nd,third Quartiles 25%,50%,75%,min,max, etc).

```python
sns.boxplot(data=df,x='total_bill',palette='Set2')
plt.title('Distribution of Total Bill ')
plt.show()
```

```python
sns.boxplot(data=df,y='total_bill',palette='Set2',fill=False)
plt.title('Distribution of Total Bill ')
plt.show()
```

```python
sns.boxplot(data=df,y='total_bill',hue='time',palette='Set1',fill=True,gap=0.2)
plt.title('Distribution of Total Bill ')
plt.show()
```

# **Density Plot**

Graphical representation of the distribution of a dataset.It uses kernel density estimation(KDE) to smoooth the data into a continuous probability density curve.Density plots are a great alternative to histogram as they provide a smoothed view of the distribution,making it easier to observe petterns without being affected bythe choice of the bin sizes.

```python
sns.histplot(data=df,x='total_bill',edgecolor='black',bins=10,palette='Set1')
plt.title("Distribution of total bill")
plt.show()
```

```python
sns.histplot(data=df,x='total_bill',edgecolor='black',bins=10,palette='Set1',element='bars')
plt.title("Distribution of total bill")
plt.show()
```

```python
sns.distplot(df['tip'],color='g')
plt.show()
```

```python
sns.distplot(df['tip'],color='g',hist=False,bins=10)
plt.show()
```

```python
sns.distplot(df['tip'],color='g',bins=10,kde=True)
plt.title("Distribution of Tip")
plt.show()
```

```python
sns.kdeplot(df['tip'],color='g')
plt.title("Distribution of Tip")
plt.show()
```

```python
sns.kdeplot(data=df,x='total_bill',hue='time',shade=True)
plt.title("Distribution of Tip")
plt.show()
```

# **Viloin Plot**

It is a data visualization that combines features of a boxplot and kernel density plot.it provides insights into the distribution of dataset by showing both summary statistics and the data density shape.

Handling Outliers.
uses for univariate analysis.
distribution of numerical variable across different categories.

```python
sns.violinplot(data=df,x='tip',palette='winter',inner_kws=dict(box_width=15,whis_width=2),split=True)
plt.title("Distribution of Tip",fontsize=18,fontweight='bold',fontname='arial',loc='left',pad=10)
plt.show()
```

# **Heatmap**

Displaying the correlation matrix of numerical values.


```python
import numpy as np
numerical=df.select_dtypes(include=np.number).columns
numerical
```

```python

numerical=df.select_dtypes(include=np.number).columns
correlation=df[numerical].corr()
correlation
```

```python
sns.heatmap(correlation,cmap='Greens',annot=True)
plt.title("Correlation between the features",fontweight='bold')
plt.show()
```

# **Combine Plot**

```python
sns.set_style("ticks")

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title("Scatter Plot of Total Bill by Tip")

plt.subplot(222)
sns.lineplot(data=df,x='gender',y='tip')
plt.title("Line Plot of Gender by Tip")

plt.subplot(223)
sns.distplot(df['total_bill'])
plt.title("Distribution  of Total Bill")

plt.subplot(224)
sns.boxplot(df['tip'])
plt.title("Distribution of Tip")

plt.tight_layout()
plt.show()
```
